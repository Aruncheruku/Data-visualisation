# Plotting Different types of plots on the climate.

# 1.Line Plot (Avg Temperature vs Year)

```python
# Plot 1: Line plot Avg Agricultural Temperature Over Time.
plt.figure(figsize=(10, 6))
avg_temp = dataset.groupby("Year")["Average_Temperature_C"].mean().reset_index()
sns.lineplot(data=avg_temp, x="Year", y="Average_Temperature_C", marker="o", linewidth=2.5, color="orange",markerfacecolor="black")
plt.title("Average Agricultural Temperature Over Time", fontsize=14,fontweight="bold")
plt.ylabel("Avg Temperature (Â°C)",fontweight="bold")
plt.xlabel("Year",fontweight="bold")
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7) 
plt.legend("Year vs Average_Temperature_C",loc = "upper left")
plt.savefig("Average Agricultural Temperature Over Time.png",dpi = 300)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 2.Box Plot(Economic Impact by Adaptation Strategy)

```python
# Plot 2: Economic Impact by Adaptation Strategy
plt.figure(figsize=(10, 8))
sns.boxplot(
    data=dataset,
    x="Adaptation_Strategies",
    y="Economic_Impact_Million_USD",
    hue="Adaptation_Strategies",   
    palette="muted",
    legend=False                         
)
plt.title("Economic Impact Across Adaptation Strategies", fontsize=12,fontweight="bold")
plt.xlabel("Adaptation Strategy",fontweight="bold")
plt.ylabel("Economic Impact (Million USD)",fontweight="bold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Economic Impact Across Adaptation Strategies.png",dpi = 300)
plt.show()
```

# 3. Heat Map

```python
# Plot 3: Correlation Heatmap of Indicators
plt.figure(figsize=(10, 8))
corr_cols = ["Average_Temperature_C", "Total_Precipitation_mm", "CO2_Emissions_MT",
             "Crop_Yield_MT_per_HA", "Extreme_Weather_Events", "Irrigation_Access_%",
             "Pesticide_Use_KG_per_HA", "Fertilizer_Use_KG_per_HA", "Soil_Health_Index",
             "Economic_Impact_Million_USD"]
corr_matrix = dataset[corr_cols].corr().round(2)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, square=True, cbar_kws={"shrink": 0.8})
plt.title("Correlation Heatmap of Climate and Agriculture Indicators", fontsize=16,fontweight="bold")

plt.tight_layout()
plt.savefig("Correlation Heatmap of Climate and Agriculture Indicators.png")
plt.show()
```
